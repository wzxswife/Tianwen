```julia
using Dates
using DataFrames
using Statistics
using LinearAlgebra

include("./scripts/TW_load.jl")
include("./scripts/TW_plot.jl")
using .TW_load
using .TW_plot
```

```julia
data_path = joinpath(@__DIR__, "..", "data", "32Hz")
file_name = "TW1_MOMAG_MSO_32Hz_20220824_2C_v03.dat"
mag2c32hz = load_mag_2c_bydlm(joinpath(data_path, file_name))
mag2c32hz[:position] .= mag2c32hz[:position] ./ 3390.0
println("Data loaded.")
```

```julia
data = mag2c32hz
shock_time = DateTime("2022-08-24T07:27:20")
time_range = shock_time - Dates.Minute(1) .+ Dates.Minute(1) .* range(0, 2)
keys_needed = [:epoch, :position, :B, :B_total]
avail_data = find_avail_data(mag2c32hz, time_range, keys_needed)
println("Available data")
```

```julia
using CairoMakie, GeometryBasics
CairoMakie.activate!()
println("Drawing...")
fig = Figure(resolution=(800, 800))
ax3 = Axis(fig[1,1], aspect=DataAspect(), xlabel=L"$x$ ($R_{\mathrm{M}}$)", 
    ylabel=L"$\sqrt{y^2+z^2}$ ($R_{\mathrm{M}}$)", xticks=range(-10, 10), yticks=range(-10, 10))
bowshock_plot_xr(ax3)
#time_range = [shock_time]
time_range = shock_time + Dates.Hour(1) .* range(0, 6)
plot_spacecraft_orbit_xr(ax3, data, time_range)
save("bowshock.png", fig)
```

```julia
using CairoMakie, GeometryBasics
CairoMakie.activate!()
println("Drawing...")
fig = Figure(resolution=(800, 1200))
ax1 = Axis(fig[1,1], aspect=DataAspect(), xlabel=L"$x$ ($R_{\mathrm{M}}$)", 
    ylabel=L"$\sqrt{y^2+z^2}$ ($R_{\mathrm{M}}$)", 
    xticks=range(-10, 10), yticks=range(-10, 10))
ax2 = Axis(fig[2,1], aspect=DataAspect(), xlabel=L"$x$ ($R_{\mathrm{M}}$)", 
    ylabel=L"$\sqrt{y^2+z^2}$ ($R_{\mathrm{M}}$)", 
    xticks=range(-10, 10), yticks=range(-10, 10))
ax3 = Axis(fig[3,1], aspect=DataAspect(), xlabel=L"$x$ ($R_{\mathrm{M}}$)", 
    ylabel=L"$\sqrt{y^2+z^2}$ ($R_{\mathrm{M}}$)", 
    xticks=range(-10, 10), yticks=range(-10, 10))
bowshock_plot(ax1, ax2, ax3)
plot_spacecraft_orbit(ax1, ax2, ax3, data, time_range)
save("bowshock.png", fig)
```
